# Why NumPy uses C
Python is slower for heavy math operations.
- So NumPy:
    - Takes your Python code
    - Sends the heavy calculations to C functions
    - Returns the result to Python

# List vs Numpy Array

```python
py_list = [1,2,3,4]
print("Python List multiplication:",py_list * 2)

np_array = np.array([1,2,3,4])
print("Python array multiplication:",np_array * 2)   #element-wise multiplication

import time
start = time.time()
py_list = [i*2 for i in range(1000000)]
print("List operation time:",time.time()-start)

start = time.time()
np_array = np.arange(1000000) * 2   # 2 is a scalar
print("Numpy operation time:",time.time()-start)
```

- np.array(range()) vs np.arange():
    - arange() stores each individual value of an array, range() stores only 3 values (start, stop, step)
    - arange() takes more space than range()
    - arange() allows floating-point numbers in step parameter, making it more flexible

# Vector, Matrix and Tensor

| Type   | Dimensions | Example shape | Real-world idea   |
| ------ | ---------- | ------------- | ----------------- |
| Vector | 1D         | `(4,)`        | List of numbers   |
| Matrix | 2D         | `(2,3)`       | Table/spreadsheet |

```python
vector = np.array([1,2,3])
print("\n",vector)

matrix = np.array([[1,2,3],[4,5,6]])
print("\n",matrix)

tensor = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print("\n",tensor)
```

# Properties of array

```python
import numpy as np
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

# shape
print("\nShape:",arr.shape)        # 2 rows 3 columns

# dimension
print("\nDimension:",arr.ndim)     # 2D array

# size (no. of elements)
print("\nSize:",arr.size)          # 6 elements

# Data type
print("\ndtype:",arr.dtype)        # experiment with different data-types however its better to keep array's homogeneous

# Itemsize (size of elements is bytes)
print("\nItem-size:",arr.itemsize)

# nbytes (total memory taken by array)
print("\nnbytes:",arr.nbytes)      # size x itemsize

# Transpose
print("\nTranspose:",arr.T)
```

## Array Reshaping

- Reshape

    -6 elements → can reshape into:
        - (2,3) ✔
        - (3,2) ✔
        - (6,1) ✔
        - (4,2) ✘ (needs 8 elements)


```python
#reshape
arr = np.array([1, 2, 3, 4, 5, 6])
new_arr = arr.reshape(2, 3)
print(new_arr)
```

- Using -1 (automatic dimension)

```python
arr = np.array([1,2,3,4,5,6])

arr.reshape(2, -1)        # prints last line because in jupiter notebook
```

- resize

```python
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

arr.resize((3,2))
print(arr)
```

- flatten

```python
#flatten
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])
arr.flatten
```

- ravel

```python
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

r = arr.ravel()     # returns view instead of copy
print(r)
```

| Feature               | `reshape()`     | `flatten()`     | `ravel()`       | `resize()`               |
| --------------------- | --------------- | --------------- | --------------- | ------------------------ |
| Purpose               | Change shape    | Make 1D copy    | Make 1D view    | Change shape of original |
| Returns copy?         | Usually no      | Yes             | No (usually)    | No                       |
| Affects original?     | No              | No              | Yes (sometimes) | Yes                      |
| Adds zeros if bigger? | No              | No              | No              | Yes                      |
| Common use            | ML data shaping | Safe flattening | Fast flattening | In-place resizing        |

# NumPy Data Types Reference Table

| Data Type | Type Code | Size | Description | Example Values |
|------------|------------|------|-------------|---------------|
| int8 | i1 | 1 byte | Signed 8-bit integer | -128 to 127 |
| uint8 | u1 | 1 byte | Unsigned 8-bit integer | 0 to 255 |
| int16 | i2 | 2 bytes | Signed 16-bit integer | -32,768 to 32,767 |
| uint16 | u2 | 2 bytes | Unsigned 16-bit integer | 0 to 65,535 |
| int32 | i4 | 4 bytes | Signed 32-bit integer | Large whole numbers |
| uint32 | u4 | 4 bytes | Unsigned 32-bit integer | Large positive whole numbers |
| int64 | i8 | 8 bytes | Signed 64-bit integer | Very large whole numbers |
| uint64 | u8 | 8 bytes | Unsigned 64-bit integer | Very large positive whole numbers |
| float16 | f2 | 2 bytes | Half-precision floating point | 3.14 |
| float32 | f4 or f | 4 bytes | Single-precision floating point | 3.1415926 |
| float64 | f8 or d | 8 bytes | Double-precision floating point (default float) | 3.141592653589793 |
| float128 | f16 or g | 16 bytes | Extended-precision floating point | High-precision decimals |
| complex64 | c8 | 8 bytes | Complex number using two float32 values | 1 + 2j |
| complex128 | c16 | 16 bytes | Complex number using two float64 values | 1 + 2j |
| complex256 | c32 | 32 bytes | Complex number using two float128 values | 1 + 2j |
| bool | ? | 1 byte | Boolean values | True, False |
| object | O | Variable | Stores any Python object | Lists, dictionaries, strings |
| string_ | S | Fixed length | ASCII string type | 'S10' |
| unicode_ | U | Fixed length | Unicode string type | 'U10' |
