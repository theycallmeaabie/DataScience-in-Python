```python
import pandas as pd
import numpy as np
```

## Correlation and Covariance

Correlation and covariance are summary statistics computed from **pairs** of values.

- Both methods automatically align values by index and skip `NaN` before computing.

### `corr` and `cov` on a Series

Called on one Series and passed another Series as an argument:

- `s1.corr(s2)` → single correlation coefficient (scalar).
- `s1.cov(s2)` → single covariance value (scalar).
- Only overlapping, non-NA, index-aligned values are used.

### `corr` and `cov` on a DataFrame

Called directly on a DataFrame with no arguments:

- `df.corr()` → full **correlation matrix** as a DataFrame.
- `df.cov()` → full **covariance matrix** as a DataFrame.
- Diagonal values of the correlation matrix are always `1.0` (a variable is perfectly correlated with itself).

### `corrwith` — Pairwise Correlation Against Another Object

Computes the correlation of each column in a DataFrame against another object:

| Argument | Returns |
|----------|---------|
| Series | Series — one correlation value per column |
| DataFrame | Series — correlation of each matching column name |

- Pass `axis="columns"` to compute row-by-row instead of column-by-column.
- Data is always aligned by label before the correlation is computed.

## Key Points

- `corr` and `cov` on a Series take another Series as an argument and return a scalar.
- `corr` and `cov` on a DataFrame return a full matrix as a DataFrame.
- All methods align on index labels and skip `NaN` automatically.
- `corrwith(series)` → correlation of each column against that Series.
- `corrwith(dataframe)` → correlation of each matching column name pair.

```python
# Synthetic stock return data (mirrors book examples using Yahoo Finance data)
np.random.seed(42)
dates = pd.date_range("2015-01-01", periods=500, freq="B")

price = pd.DataFrame(
    np.random.uniform(50, 200, size=(500, 4)).cumprod(axis=0) / 1e10,
    index=dates,
    columns=["AAPL", "GOOG", "IBM", "MSFT"]
)

volume = pd.DataFrame(
    np.random.randint(10_000_000, 100_000_000, size=(500, 4)),
    index=dates,
    columns=["AAPL", "GOOG", "IBM", "MSFT"]
)

returns = price.pct_change()

returns.tail()
```

```python
# Series corr and cov: scalar result
print(returns["MSFT"].corr(returns["IBM"]))
print(returns["MSFT"].cov(returns["IBM"]))
```

```python
returns.corr()  # full correlation matrix
```

```python
returns.cov()  # full covariance matrix
```

```python
returns.corrwith(returns["IBM"])  # each column correlated against IBM
```

```python
returns.corrwith(volume)  # matching column names correlated between two DataFrames
```
