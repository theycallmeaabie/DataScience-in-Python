## Summarizing and Computing Descriptive Statistics

pandas objects have built-in mathematical and statistical methods that fall into two main categories:

- **Reductions / summary statistics** — extract a single value from a Series, or a Series of values from a DataFrame's rows or columns (e.g. `sum`, `mean`).
- **Accumulations** — compute a running total or running result (e.g. `cumsum`).

A key advantage over NumPy: pandas methods have **built-in handling for missing data (`NaN`)**.

### NaN Handling

- By default, `NaN` values are **skipped** (`skipna=True`).
- A column or row of **all `NaN`** values:
  - `sum` → `0`
  - `mean` → `NaN` (requires at least one non-NA value)
- Set `skipna=False` to propagate `NaN`: any `NaN` in a group makes the result `NaN`.

### Reduction Method Options

| Option | Description |
|--------|-------------|
| `axis` | Axis to reduce over: `"index"` (across rows) or `"columns"` (across columns) |
| `skipna` | Exclude `NaN` values; `True` by default |
| `level` | Reduce grouped by level if the axis is a `MultiIndex` |

### Types of Methods

| Type | Examples | Returns |
|------|----------|---------|
| Reductions | `sum`, `mean`, `min`, `max`, `count` | Scalar or Series |
| Indirect statistics | `idxmin`, `idxmax` | Index label of min/max value |
| Accumulations | `cumsum`, `cummin`, `cummax`, `cumprod` | Same shape as input |
| Summary | `describe` | Multiple statistics at once |

### `describe`

- On **numeric** data: returns count, mean, std, min, quartiles, max.
- On **non-numeric** data: returns count, unique, top (most frequent), freq.

### Full Reference

| Method | Description |
|--------|-------------|
| `count` | Number of non-NA values |
| `describe` | Set of summary statistics |
| `min`, `max` | Minimum and maximum values |
| `argmin`, `argmax` | Integer index location of min/max (Series only) |
| `idxmin`, `idxmax` | Index label of min/max value |
| `quantile` | Sample quantile from 0 to 1 (default 0.5) |
| `sum` | Sum of values |
| `mean` | Mean of values |
| `median` | Arithmetic median (50th percentile) |
| `mad` | Mean absolute deviation from mean |
| `prod` | Product of all values |
| `var` | Sample variance |
| `std` | Sample standard deviation |
| `skew` | Sample skewness (third moment) |
| `kurt` | Sample kurtosis (fourth moment) |
| `cumsum` | Cumulative sum |
| `cummin`, `cummax` | Cumulative minimum or maximum |
| `cumprod` | Cumulative product |
| `diff` | First arithmetic difference (useful for time series) |
| `pct_change` | Percent changes |

## Key Points

- pandas statistical methods skip `NaN` by default (`skipna=True`).
- Use `axis="columns"` to compute across columns (per row); default is `axis="index"` (per column).
- An all-`NaN` column/row gives `0` for `sum` but `NaN` for `mean`.
- `skipna=False` makes any `NaN` in the group propagate to the result.
- `idxmin`/`idxmax` return the **label** of the min/max, not the value itself.
- `describe` gives a full statistical summary; output varies for numeric vs non-numeric data.

```python
df = pd.DataFrame(
    [[1.4, np.nan], [7.1, -4.5], [np.nan, np.nan], [0.75, -1.3]],
    index=["a", "b", "c", "d"],
    columns=["one", "two"]
)

df
```

```python
df.sum()  # sum per column (NaN skipped)
```

```python
df.sum(axis="columns")  # sum per row
```

```python
df.sum(axis="index", skipna=False)  # NaN propagates per column
```

```python
df.sum(axis="columns", skipna=False)  # NaN propagates per row
```

```python
df.mean(axis="columns")  # all-NaN row yields NaN
```

```python
df.idxmax()  # index label of max value per column
```

```python
df.cumsum()
```

```python
df.describe()
```

```python
obj = pd.Series(["a", "a", "b", "c"] * 4)

obj.describe()  # non-numeric: count, unique, top, freq
```
