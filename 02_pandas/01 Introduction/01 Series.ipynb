# Introduction to pandas Data Structures

Pandas provides two primary data structures:

- Series: One-dimensional labeled array
- DataFrame: Two-dimensional labeled table

## Series

A Series is a one-dimensional array-like object containing:

- A sequence of values
- An associated index (labels)

If no index is provided, pandas creates a default integer index from 0 to N-1.

## Custom Index

A custom index can be provided to label each value.

## Selecting Data

Unlike NumPy arrays, Series values can be accessed using labels.

## Boolean Filtering

Series supports Boolean filtering similar to NumPy arrays.

## Arithmetic Operations

Arithmetic operations are applied element-wise while preserving labels.

## NumPy Functions

NumPy functions work directly on Series and preserve the index.

## Key Points

- Series is a 1D labeled array.
- Every Series contains values and an index.
- Custom labels can be assigned using the index parameter.
- Values can be accessed using labels.
- Arithmetic operations preserve labels.
- NumPy functions work seamlessly with Series.
- Boolean filtering works similarly to NumPy arrays.

```python
import pandas as pd

obj = pd.Series([4, 7, -5, 3])
```

```python
obj2 = pd.Series(
    [4, 7, -5, 3],
    index=["d", "b", "a", "c"]
)
```

```python
obj2["a"]
```

```python
obj2[obj2 > 0]
```

```python
obj2 * 2
```

## Series as a Dictionary

A pandas Series can be thought of as a **fixed-length, ordered dictionary** because it maps index labels to values.

### Membership Testing

Like a dictionary, you can check whether a label exists in a Series using the `in` operator.

### Creating a Series from a Dictionary

A Series can be created directly from a Python dictionary. The dictionary keys become the index, and the dictionary values become the Series values.

### Converting Back to a Dictionary

Use the `to_dict()` method to convert a Series back into a Python dictionary.

### Specifying a Custom Index

When creating a Series from a dictionary, you can provide a custom index.

- Keys found in the dictionary are included.
- Missing keys are assigned `NaN` (Not a Number).
- Dictionary keys not present in the custom index are excluded.

### Missing Values (`NaN`)

`NaN` represents missing data in pandas.

In the example:

- `"California"` is included in the index but not in the dictionary → `NaN`
- `"Utah"` exists in the dictionary but not in the custom index → excluded

## Key Points

- A Series behaves similarly to an ordered dictionary.
- Dictionary keys become Series indices.
- `to_dict()` converts a Series back to a dictionary.
- Custom indices control which labels appear and in what order.
- Missing values are represented by `NaN`.

```python
"b" in obj2
```

```python
"e" in obj2
```

```python
sdata = {
    "Ohio": 35000,
    "Texas": 71000,
    "Oregon": 16000,
    "Utah": 5000
}

obj3 = pd.Series(sdata)
```

```python
obj3.to_dict()
```

```python
states = ["California", "Ohio", "Oregon", "Texas"]

obj4 = pd.Series(sdata, index=states)
```

## Detecting Missing Data

In pandas, the terms:

- Missing
- NA
- Null

are used interchangeably to refer to missing values.

Missing values are typically represented as `NaN` (Not a Number).

### Detecting Missing Values

Use:

- `pd.isna()` → returns `True` for missing values.
- `pd.notna()` → returns `True` for non-missing values.

### Instance Methods

Series objects also provide:

- `isna()`
- `notna()`

These methods behave the same as the corresponding pandas functions.

## Key Points

- Missing values in pandas are represented by `NaN`.
- `isna()` identifies missing values.
- `notna()` identifies non-missing values.
- Both pandas functions and Series methods are available.
- Results are returned as Boolean Series.
