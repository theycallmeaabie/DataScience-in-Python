```python
import pandas as pd
import numpy as np
```

## pandas Index Objects

An **Index** object stores the axis labels of a Series or DataFrame.

Examples:

- Series index labels
- DataFrame row labels
- DataFrame column names

Whenever you provide labels while creating a Series or DataFrame, pandas automatically converts them into an `Index` object.

---

## Characteristics of Index

### Array-like

An Index behaves similarly to an array:

- Supports indexing
- Supports slicing
- Stores labels

### Immutable

Index objects cannot be modified after creation.

This prevents accidental changes and makes them safe to share between multiple pandas objects.

---

## Shared Index Objects

Multiple Series or DataFrames can share the same Index object.

Because Index objects are immutable, sharing them is safe and memory-efficient.

---

## Index as a Set

Besides being array-like, an Index also behaves somewhat like a fixed-size set.

You can test membership using the `in` operator.

Examples:

- Check if a column exists.
- Check if a row label exists.

---

## Duplicate Labels

Unlike Python sets:

- Index objects can contain duplicate labels.

This is useful when working with real-world datasets that may have repeated identifiers.

When selecting a duplicate label:

- All matching entries are returned.

## Key Points

- Index stores row and column labels.
- Labels provided during object creation become an Index.
- Index objects are immutable.
- Multiple objects can safely share the same Index.
- Membership testing is supported with `in`.
- Duplicate labels are allowed.
- Selecting a duplicate label returns all matching entries.

```python
obj = pd.Series(
    np.arange(3),
    index=["a", "b", "c"]
)

index = obj.index

obj
```

```python
index[1:]
# index[1] = 'd' => TypeError
```

```python
labels = pd.Index(np.arange(3))

obj2 = pd.Series(
    [1.5, -2.5, 0],
    index=labels
)

obj2
```

```python
obj2.index is labels
```

## Common Index Methods and Properties

Pandas Index objects provide many useful methods for set operations, searching, and modifying labels.

### Set Operations

| Method | Description |
|----------|-------------|
| `append()` | Concatenate with another Index |
| `difference()` | Values present in first Index but not second |
| `intersection()` | Common values between Index objects |
| `union()` | All unique values from both Index objects |

### Membership Operations

| Method | Description |
|----------|-------------|
| `isin()` | Returns a Boolean array indicating membership |

### Modification Operations

Since Index objects are immutable, these methods return a **new Index** rather than modifying the original.

| Method | Description |
|----------|-------------|
| `delete()` | Remove element at a given position |
| `drop()` | Remove specified labels |
| `insert()` | Insert a value at a given position |

### Properties

| Property | Description |
|-----------|-------------|
| `is_monotonic_increasing`, `is_monotonic_decreasing` | True if values are sorted in ascending order , descending order respectively |
| `is_unique` | True if no duplicate labels exist |

### Unique Values

| Method | Description |
|----------|-------------|
| `unique()` | Returns unique labels in the Index |

## Key Points

- Index supports set-like operations.
- Index objects are immutable.
- Methods like `delete`, `drop`, and `insert` return new Index objects.
- `is_monotonic` checks sorted order.
- `is_unique` checks for duplicates.
- `unique()` returns distinct labels.

```python
idx1 = pd.Index(["a", "b", "c"])
idx2 = pd.Index(["b", "c", "d"])

idx1.union(idx2)
```

```python
idx1.intersection(idx2)
```

```python
idx1.difference(idx2)
```

```python
idx1.isin(["a", "c"])
```

```python
idx1.delete(1)
```

```python
idx1.insert(1, "x")
```

```python
pd.Index([1, 2, 3, 4]).is_monotonic_increasing
```

```python
pd.Index(["a", "b", "c"]).is_unique
```

```python
pd.Index(["a", "a", "b", "b"]).unique()
```
