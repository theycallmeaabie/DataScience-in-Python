## Axis Indexes with Duplicate Labels

pandas does **not** require index labels to be unique — duplicates are permitted.

However, some operations (like `reindex`) do require unique labels.

### Checking for Uniqueness

Use the `is_unique` property on an index to check whether all labels are unique:

```python
obj.index.is_unique  # True or False
```

### Selection Behavior with Duplicate Labels

The return type of label-based indexing depends on whether the label is duplicated:

| Label in index | `obj[label]` returns |
|----------------|----------------------|
| Unique (appears once) | Scalar value |
| Duplicate (appears multiple times) | Series of all matching values |

This inconsistency in return type can complicate downstream code — a function expecting a scalar may unexpectedly receive a Series.

The same behavior applies to **DataFrame** row (and column) selection with `loc`:

- Duplicate row label → returns a DataFrame.
- Unique row label → returns a Series.

## Key Points

- Duplicate index labels are allowed but not required.
- `index.is_unique` returns `True` if all labels are distinct.
- Indexing a **unique** label returns a scalar (or Series for a DataFrame row).
- Indexing a **duplicate** label returns a Series (or DataFrame for a DataFrame row).
- The variable return type is a potential source of bugs when labels may or may not be duplicated.

````python
obj = pd.Series(np.arange(5), index=["a", "a", "b", "b", "c"])

obj
````

````python
obj.index.is_unique
````

````python
obj["a"]  # duplicate label → returns a Series
````

````python
obj["c"]  # unique label → returns a scalar
````

````python
df = pd.DataFrame(
    np.random.standard_normal((5, 3)),
    index=["a", "a", "b", "b", "c"]
)

df
````

````python
df.loc["b"]  # duplicate row label → returns a DataFrame
````

````python
df.loc["c"]  # unique row label → returns a Series
````
