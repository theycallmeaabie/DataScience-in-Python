```python
import pandas as pd
import numpy as np
```

## Dropping Entries from an Axis

The `drop` method returns a **new object** with the specified labels removed from an axis.

- The original object is not modified.
- A single label or a list of labels can be dropped.

### On a Series

Pass one label or a list of labels to remove them from the index.

### On a DataFrame

Labels can be dropped from either axis:

| Approach | Drops from |
|----------|------------|
| `drop(index=[...])` | Row labels (axis 0) |
| `drop(columns=[...])` | Column labels (axis 1) |
| `drop([...], axis=1)` | Column labels (axis 1) — NumPy-style |
| `drop([...], axis="columns")` | Column labels (axis 1) — named-axis style |

## Key Points

- `drop` returns a new object — it does not modify in place.
- A single label or a list of labels can be passed.
- For DataFrames, use `index=` to drop rows and `columns=` to drop columns.
- `axis=1` and `axis="columns"` are equivalent alternatives for dropping columns.

```python
obj = pd.Series(np.arange(5.), index=["a", "b", "c", "d", "e"])

obj
```

```python
new_obj = obj.drop("c")

new_obj
```

```python
obj.drop(["d", "c"])
```

```python
data = pd.DataFrame(
    np.arange(16).reshape((4, 4)),
    index=["Ohio", "Colorado", "Utah", "New York"],
    columns=["one", "two", "three", "four"]
)

data
```

```python
data.drop(index=["Colorado", "Ohio"])
```

```python
data.drop(columns=["two"])
```

```python
data.drop("two", axis=1)
```

```python
data.drop(["two", "four"], axis="columns")
```
