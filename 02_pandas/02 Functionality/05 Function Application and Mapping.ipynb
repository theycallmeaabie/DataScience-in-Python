```python
import pandas as pd
import numpy as np
```

## NumPy ufuncs on pandas Objects

NumPy universal functions (ufuncs) operate **element-wise** and work directly on Series and DataFrames.

- Labels and index are preserved in the result.
- No special setup required — pass a pandas object where you would pass a NumPy array.

## Key Points

- NumPy ufuncs (e.g. `np.abs`, `np.sqrt`, `np.exp`) work element-wise on pandas objects.
- Index and column labels are retained in the output.

```python
frame = pd.DataFrame(
    np.random.standard_normal((4, 3)),
    columns=list("bde"),
    index=["Utah", "Ohio", "Texas", "Oregon"]
)

frame
```

```python
np.abs(frame)
```

## `apply` — Applying a Function Along an Axis

`apply` applies a function to **each column or each row** of a DataFrame.

### Axis Behavior

| Argument | Function receives | Applied to |
|----------|-------------------|------------|
| `axis="index"` (default) | Each column as a Series | Per column |
| `axis="columns"` | Each row as a Series | Per row |

A helpful way to think about `axis="columns"`: the function is applied **across the columns** of each row.

### Return Value

- Function returns a **scalar** → result is a Series (one value per column or row).
- Function returns a **Series** → result is a DataFrame (each returned Series becomes a column).

### Note

Common reductions like `sum` and `mean` are built-in DataFrame methods — `apply` is not needed for those.

## Key Points

- `apply` passes each column (or row) as a Series to the function.
- Default axis is `"index"` → operates column-by-column.
- `axis="columns"` → operates row-by-row.
- Scalar return → Series result; Series return → DataFrame result.

```python
def f1(x):
    return x.max() - x.min()

frame.apply(f1)  # applied per column
```

```python
frame.apply(f1, axis="columns")  # applied per row
```

```python
def f2(x):
    return pd.Series([x.min(), x.max()], index=["min", "max"])

frame.apply(f2)  # function returns a Series → result is a DataFrame
```

## Element-Wise Mapping: `applymap` and `map`

For applying a function to **every individual element**, pandas provides:

| Method | Object | Applies to |
|--------|--------|------------|
| `applymap(func)` | DataFrame | Every element in the DataFrame |
| `map(func)` | Series | Every element in the Series |

- `applymap` is the DataFrame equivalent of Series's `map`.
- Both return an object of the same shape with the function applied to each value.

## Key Points

- `applymap` applies an element-wise function across an entire DataFrame.
- `map` applies an element-wise function to every value in a Series.
- Both preserve the shape and labels of the original object.

```python
def my_format(x):
    return f"{x:.2f}"

frame.applymap(my_format)
```

```python
frame["e"].map(my_format)
```
