```python
import pandas as pd
import numpy as np
```

## Reindexing

`reindex` creates a **new object** with values rearranged to align with a new index.

- The original object is not modified.
- If a new index label did not exist in the original, its value becomes `NaN`.
- If a label from the original is not in the new index, that data is dropped.

### On a Series

Passing a new list of labels to `reindex` reorders the data to match:

- Existing labels → values carried over.
- New labels not in the original → `NaN`.

### Filling / Interpolation with `method`

Useful for ordered data such as time series where gaps should be filled.

| Method | Behavior |
|--------|----------|
| `ffill` | Forward-fill: propagates the last valid value forward |
| `bfill` | Backward-fill: propagates the next valid value backward |

### On a DataFrame

`reindex` can alter the row index, column labels, or both.

- Passing a sequence as a positional argument → reindexes **rows**.
- Using `index=` keyword → reindexes **rows**.
- Using `columns=` keyword → reindexes **columns**.
- Using `axis=` keyword with a label sequence → specify which axis to reindex.

### Alternative: `loc`

You can also reindex using `loc`, which many users prefer.

- Works only if **all** new labels already exist in the object.
- Unlike `reindex`, it will not insert `NaN` for missing labels.

## Key Points

- `reindex` creates a new object — it does not modify in place.
- Missing labels introduced by reindexing are filled with `NaN` by default.
- Labels not included in the new index are dropped.
- `method='ffill'` forward-fills gaps; `method='bfill'` backward-fills.
- For DataFrames, use `columns=` to reindex columns and `index=` (or a positional sequence) for rows.
- `axis='columns'` can also be used to target the column axis.
- `loc` is an alternative that works only when all labels already exist.

```python
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=["d", "b", "a", "c"])

obj
```

```python
obj2 = obj.reindex(["a", "b", "c", "d", "e"])

obj2
```

```python
obj3 = pd.Series(["blue", "purple", "yellow"], index=[0, 2, 4])

obj3
```

```python
obj3.reindex(np.arange(6), method="ffill")
```

```python
frame = pd.DataFrame(
    np.arange(9).reshape((3, 3)),
    index=["a", "c", "d"],
    columns=["Ohio", "Texas", "California"]
)

frame
```

```python
frame2 = frame.reindex(index=["a", "b", "c", "d"])

frame2
```

```python
states = ["Texas", "Utah", "California"]

frame.reindex(columns=states)
```

```python
frame.reindex(states, axis="columns")
```

```python
# loc alternative: only works when all labels already exist
frame.loc[["a", "d", "c"], ["California", "Texas"]]
```
