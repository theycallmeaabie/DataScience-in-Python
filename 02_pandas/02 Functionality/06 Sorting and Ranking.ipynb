```python
import pandas as pd
import numpy as np
```

## Sorting by Index — `sort_index`

`sort_index` sorts a Series or DataFrame **lexicographically by its labels** and returns a new sorted object.

- The original object is not modified.
- Default sort order is **ascending**.
- Pass `ascending=False` for descending order.

### On a DataFrame

| Argument | Sorts |
|----------|-------|
| `sort_index()` | Row labels (axis 0) |
| `sort_index(axis="columns")` | Column labels (axis 1) |

## Key Points

- `sort_index` sorts by **labels**, not values.
- Returns a new object — does not modify in place.
- Default is ascending; use `ascending=False` for descending.
- For DataFrames, use `axis="columns"` to sort column labels.

```python
obj = pd.Series(np.arange(4), index=["d", "a", "b", "c"])

obj
```

```python
obj.sort_index()
```

```python
frame = pd.DataFrame(
    np.arange(8).reshape((2, 4)),
    index=["three", "one"],
    columns=["d", "a", "b", "c"]
)

frame
```

```python
frame.sort_index()
```

```python
frame.sort_index(axis="columns")
```

```python
frame.sort_index(axis="columns", ascending=False)
```

## Sorting by Values — `sort_values`

`sort_values` sorts by the **data values** rather than the index labels.

### On a Series

- Sorts by value; the index follows along.
- `NaN` values are placed at the **end** by default.
- Use `na_position="first"` to place `NaN` at the start.

### On a DataFrame

- Pass a column name (or list of column names) as the sort key.
- When sorting by multiple columns, the first column is the primary key; subsequent columns break ties.

## Key Points

- `sort_values` sorts by data values, not labels.
- `NaN` is placed at the end by default; use `na_position="first"` to move it to the start.
- For DataFrames, pass one column name or a list of column names as sort keys.
- Earlier columns in the list take priority when sorting by multiple columns.

```python
obj = pd.Series([4, 7, -3, 2])

obj.sort_values()
```

```python
obj = pd.Series([4, np.nan, 7, np.nan, -3, 2])

obj.sort_values()
```

```python
obj.sort_values(na_position="first")
```

```python
frame = pd.DataFrame({"b": [4, 7, -3, 2], "a": [0, 1, 0, 1]})

frame
```

```python
frame.sort_values("b")
```

```python
frame.sort_values(["a", "b"])  # sort by "a" first, then "b" to break ties
```

## Ranking — `rank`

`rank` assigns a rank to each value from **1** (lowest) to **n** (highest number of valid data points).

- By default, tied values are assigned their **average rank**.
- `NaN` values are excluded from ranking.
- Pass `ascending=False` to rank from highest to lowest.

### Tie-Breaking Methods

| `method` | Behavior |
|----------|----------|
| `"average"` | Default — assign the average rank to each tied entry |
| `"min"` | Use the minimum rank for all tied entries |
| `"max"` | Use the maximum rank for all tied entries |
| `"first"` | Assign ranks in the order the values appear in the data |
| `"dense"` | Like `"min"`, but ranks always increase by 1 between groups |

### On a DataFrame

- `rank()` ranks within each column by default (axis 0).
- `rank(axis="columns")` ranks each value relative to the other values in its row.

## Key Points

- `rank` assigns ordinal ranks starting from 1.
- Ties are broken with `method="average"` by default.
- `ascending=False` ranks the largest value as 1.
- For DataFrames, use `axis="columns"` to rank across each row.

```python
obj = pd.Series([7, -5, 7, 4, 2, 0, 4])

obj.rank()  # tied values get average rank
```

```python
obj.rank(method="first")  # tied values ranked by order of appearance
```

```python
obj.rank(ascending=False)  # highest value gets rank 1
```

```python
frame = pd.DataFrame({
    "b": [4.3, 7, -3, 2],
    "a": [0, 1, 0, 1],
    "c": [-2, 5, 8, -2.5]
})

frame
```

```python
frame.rank(axis="columns")  # rank each value relative to others in its row
```
